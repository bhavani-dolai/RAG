```python
import ollama
def ask_llm(inp):
    response=ollama.chat(
        model="llama3:8b",
        messages=[{
            "role":"user",
            "content":inp
        }],
        stream=True
    )
    for res in response:
         print(res["message"]["content"],end="",flush=True)
    
```

```python
while True:
    inp=input("Enter the input")
    ask_llm(inp)
```

```python
from langchain_docling.loader import DoclingLoader

FILE_PATH = r"C:/Users/golve/Downloads/JupyterLab_Theory_Document_3_Pages.pdf"

loader = DoclingLoader(file_path=FILE_PATH)
docs = loader.load()

print(docs)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter 
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 200 , chunk_overlap = 60)
texts = text_splitter.split_documents(docs)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name = "BAAI/bge-large-en-v1.5")
```

```python
from langchain_chroma import Chroma

vector_store = Chroma(
    collection_name = "Sample_RAG",
    embedding_function = embeddings,
    persist_directory = "./chroma_langchain_db",
)
```

```python
vector_store._collection.count()
```

```python
from langchain_community.vectorstores.utils import filter_complex_metadata
texts=filter_complex_metadata(texts)
```

```python
texts[10]
```

```python
vector_store.add_documents(texts)
```

```python
vector_store._collection.count()
```

```python
vector_store._collection.get(ids=['c89b2a6e-e4a1-4b62-9a9e-460435674645'],include=["embeddings","documents"])
```

```python
retriever = vector_store.as_retriever()
```

```python
query="what are applications of jupyterlab"
retrieved_docs=retriever.invoke(query)
context="\n\n".join(
    [doc.page_content for doc in retrieved_docs]
)
print(context)

prompt = f"""
You are a Intelligent RAG assistant and you need to answer quries based on provided context.
If the answer is not present in the provided context then politely return "I donot know"
The input is {query}
The retrived context is {context}"""
ask_llm(prompt)
```
